```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
def remove_columns(df, threshold=0.9):

    # Calculate the proportion of non-zero values for each column
    non_zero_ratio = (df != 0).mean()

    # Step 2: Filter out columns that have a non-zero ratio below the threshold
    columns_to_keep = non_zero_ratio[non_zero_ratio > threshold].index
    df_all_filtered = df[columns_to_keep]

    # Display the updated lists and the filtered DataFrame
    print("Original DataFrame Shape:", df.shape)
    print("Filtered DataFrame Shape:", df_all_filtered.shape)
    
    return df_all_filtered
```

```python
def plot_histogram(df, title):
    # Plot histograms for each numeric feature
    df.hist(bins=20, figsize=(18, 12))
    plt.suptitle(title, fontsize=16)  # Title for class features
    plt.tight_layout(rect=[0, 0.03, 1, 0.95])  # Adjust layout
    plt.show()
```

```python
def train_test_split_df(df, train_size=0.9):
    X_train, X_test = train_test_split(df, train_size=train_size, random_state=42)
    
    df_train = pd.DataFrame(X_train, columns=df.columns)
    df_test = pd.DataFrame(X_test, columns=df.columns)
    
    return df_train, df_test
```
